noticing the above -> as threshold 0.5-> 0.1 , recall 0.88->0.92 (~4%) and precision 6.9->1.5 (~5%) ....this means it is not a very clever choice to use this model cause setting the threshold to 0.1 or less than 0.5 would mean that almost every alert is false (cause we are generating alot of fp to make the recall high) and setting the threshold >0.5 would make us miss the fraud cases as recall drops which is very expensive -> switch to some other model -> decision tree classifier
Conclusion : model is not bad but threshold tuning is not enough to solve its precision-recall tradeoff.
